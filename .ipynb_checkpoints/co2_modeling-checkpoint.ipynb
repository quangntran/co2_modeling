### Linear 

```python
# fit linear function to the data
# code from https://towardsdatascience.com/polynomial-regression-bbe8b9d97491
def fit_poly(degree, x, y):
    """
    Fit polynomial function with a certain degree to the data,
    and return the predictions for t
    """
    polynomial_features= PolynomialFeatures(degree=degree)
    x_poly = polynomial_features.fit_transform(x)

    model = LinearRegression()
    model.fit(x_poly, y)
    print('Coef:', model.coef_)
    print('Intercept', model.intercept_)
    y_poly_pred = model.predict(x_poly)
    return y_poly_pred

# fit
# To pass as arguments to y_poly_pred, we need to transform the data to 
# include another axis (hence the [:, np.newaxis])
y_poly_pred1 = fit_poly(1, x=np.array(train_dates)[:, np.newaxis], y=np.array(train_co2)[:, np.newaxis])
```

```python
# plot the fit line
plt.plot(train_dates, train_co2)
plt.plot(train_dates, y_poly_pred1, color='red', label='Best fit line')
plt.xlabel('Days from 1958-03-29')
plt.ylabel('CO2 concentration (ppm)')
plt.legend()
```

```python
# Calculate MSE
def mse(y,x):
    """
    y and x are numpy arrays of the same length
    """
    return(np.sum(y-x)**2)/len(y)
mse(y_poly_pred1, np.array(train_co2))
# np.sum((y_poly_pred - np.array(train_co2))**2)
```

We see that there is a "trend" in the supposed seasonality component itself (a wave like function with period of around 10,000). We don't want this, because it has to be just a seasonal component center around 0.

### A proposed solution: use a piecewise function for the trend

We observe that there seems like a sudden rise and drop around 12150. We will split data at this pivot. 

```python
## looking for the index of points closest to 12150
critical_ind = np.argmin((np.array(train_dates) - 12150)**2)
# split into two parts
first_dates, first_co2 = train_dates[:critical_ind], train_co2[:critical_ind]
second_dates, second_co2 = train_dates[critical_ind:], train_co2[critical_ind:]
## fitting quadratic to the first part
y_first_pred = fit_poly(degree=2, x=np.array(first_dates)[:, np.newaxis], 
                       y=np.array(first_co2)[:, np.newaxis])

## fitting quadratic to the second part
y_second_pred = fit_poly(degree=2, x=np.array(second_dates)[:, np.newaxis], 
                       y=np.array(second_co2)[:, np.newaxis])
plt.plot(train_dates, train_co2)
plt.plot(first_dates, y_first_pred, color='red', label='Best fit curve')
plt.plot(second_dates, y_second_pred, color='red')
plt.xlabel('Days from 1958-03-29')
plt.ylabel('CO2 concentration (ppm)')
# plt.xlim(9000,18000)
plt.legend()
```

```python
# calculate the MSE, lower than that of quadratic function
mse(np.array(list(y_first_pred[:,0])+list(y_second_pred[:,0])), np.array(train_co2))
```

### Function for the asymmetric triangle wave function


![fsd]("./img/img")

Because of the lopsided feature, I will choose an asymmetric triangle periodic function. 

The idea to build the function is inspired by this thread: https://stackoverflow.com/a/41906140

This function fluctuates around 0, and has the symmetric amplitude (above and below the 0 horizontal line)

```python
def periodic(x, phase, T, amin, amax, rising_t):
    localT = ((x-phase) % T) 
    if localT <= rising_t:
        return amin + (amax - amin) * (localT/rising_t)
    else:
        return amax - (amax - amin) * (localT - rising_t) / (T - rising_t)

# demo
x_  = np.linspace(-10,20,1000)
y_ = []
for item in x_:
    y_.append(periodic(item,phase=-1, T=6, amin=-2, amax=2, rising_t=2))
plt.plot(x_,y_)
plt.title('Asymmetric triangle periodic function with $\phi=-1, T=6, A=2$, rising period=2')
```

# The model

Let $T=365.25$

For $t<12150$:
$$p(x_t |\theta)=N(c_1+c_2t+c_3t^2+\delta(t;A,T,\phi,r),\sigma^2)$$
For $t\ge12150$:
$$p(x_t |\theta)=N(c_4+c_5t+c_6t^2+\delta(t;A,T,\phi,r),\sigma^2)$$
where $\delta(t;A,T,\phi,r)$ is the periodic function:

For $(t-\phi) \text{mod}(T) \le r$:
$$\delta(t;A,T,\phi,r)=-A + 2A \frac{(t-\phi) \text{mod}(T)}{r}$$
For $(t-\phi) \text{mod}(T) > r$:
$$\delta(t;A,T,\phi,r)=A-2A\frac{(t-\phi) \text{mod}(T)-r}{T-r}$$

### Constraints

The constraints for parameters of the seasonlity function $\delta$:

1. $A$: the amplitude, positive ($A>0$)
2. $\phi$: the phase. If we look at the image above on an example of the asymmetric periodic function, we see that if the phase $\phi$ has an effect of shifting the entire graph horizontally to the right $\phi$ units (a negative value of $\phi$ means shifting to the left. Therefore, we constraint $\phi$ to be positive and less than the period $T$ ($0 \le \phi \le T$)
3. $r$: the rising period. Of the period $T$, the graph can rise for $r$ units. If $r=T$, then it's the function is the sawtooth function. $0<r<T$

The constraints for parameters associated with the noise:

1. $\sigma$: positive value ($\sigma >0$)

The constraints for parameters associated with the trend:

This is the part I'm least certain about. The constraints should just be on $c_3$ and $c_6$ ($c_3, c_6 > 0$) because they are coefficients of the leading terms in the quadratic functions. However, if I put constraints on  only $c_3, c_6$, the model results are bad ($\hat{R}$ is large, chains don't mix, etc.) Only when I put the same constraints on $c_2, c_5 $ ($c_2,c_5>0$) does the model converge. I admit that I have no good reason for putting these extra constraints. Regardless, I will enumerate the constraints below.

1. $c_1, c_4$: real values
2. $c_2, c_3, c_5, c_6>0$



```python
# Hyperparameter Selection
# find hyperparameters 
from scipy.optimize import minimize
import scipy
import scipy.stats as sts
def f(x, lo, hi):
    alpha = x[0]
    beta = x[1]
    lower = sts.gamma(a=alpha, scale=1/beta).cdf(lo)
    upper = sts.gamma(a=alpha, scale=1/beta).cdf(hi)
    return (lower-0.025)**2 + (upper-0.975)**2 + ((hi+lo)/2 * beta + 1 - alpha)**2

# params for gamma A
x_initial = [1,1]
result = minimize(f, x_initial, (.05,.5))
x_final = result.x
x_final
```

### Choice of priors

Priors for patameters of the seasonal component:

1. $A$: I believe the amplitude should not exceed 0.5, and should not go below 0.05 (on the normalized scale). Additionally, because $A>0$, I will choose gamma distribution $gamma(\alpha=11.38643189, \beta=37.76884397)$. The parameters of the gamma are chosen that way so that the 95% confidence intreval is roughly [0.05, 0.5] and the mean is $(0.5+0.05)/2$
2. $\phi$: It can't exceed T, but can't go below 0. A uniform prior from 0 to T/2 would be approrpiate as it's uninformative. However, when I ran Stan with the uniform prior, it does not converge. Therefore, I experimented with other distributions and chose another distribution, Exponential(2/T), which makes Stan run well. Exponential(2/T) is still acceptable, because: 

    * Its mean is T/2. Because we don't know the phase, and we know its range is from 0 to T, having the distribution center around the middle point of the range is sensible.
    * 2/T is a relatively small number, being able to reflect our uncertainty. (In exponential distribution, small lambda equals large variance/uncertainty)
    
3. $r$: Because its range is T, we may put a uniform prior over [0, T/2]. Due to the same reason with $\phi$, I chose exponential(2/T).

Priors for parameters of the trend component:

1. $c_1,c_4$: The only constraints these have are real values. Because in the scale of normalized data, the data for CO2 level only in the scale of $1e-1$, it's unlikely to have the coefficient larger than 100 times the scale. Therefore, I choose Normal(0, 10).
2. $c_2, c_3, c_5, c_6$: These are real positive values, I planned to put a broad prior on them, $Exp(1e-6)$ (in exponential, small lambda equals large variance/uncertainty). However, when I did that, Stan would not converge. I tried changing the priors for $c_2$ and $c_5$ to Normal(0,10) and it works just fine. I did not know that Stan's performance has such a dependence on the choice of priors. I don't know what about the exponential distribution that makes Stan do not work well. As such:
    * Priors for $c_3, c_6:$ Exponential(1e-6)
    * Priors for $c_2, c_5$: Normal(0,10)
    

Prior for sigma of noise:

1. $\sigma:$ Exponential(1e-6), for the same reason of choosing exponential(1e-6) above.

```python
stan_code = """

// The data block contains all known quantities - typically the observed
// data and any constant hyperparameters.
data {
    int<lower=1> num;  // number of data in the data set
    real<lower=0> T; // period of seasonality, equals 365.25
    real<lower=0> t[num]; // days since 1958-03-29
    real co2[num]; // co2 level

}

// The parameters block contains all unknown quantities - typically the
// parameters of the model. Stan will generate samples from the posterior
// distributions over all parameters.
parameters {
    real c1; //coefficient in quadratic function 
    real<lower=0> c2; //coefficient in quadratic function 
    real<lower=0> c3; //coefficient in quadratic function 
    real c4; //coefficient in quadratic function 
    real<lower=0> c5; //coefficient in quadratic function 
    real<lower=0> c6; //coefficient in quadratic function 
    real<lower=0> A; // amplitude of the periodic function
    real<lower=0, upper=T> phi; // phase
    real<lower=0, upper=T> r; // rising time
    real<lower=0> sig; // std of the normal noise
    
    
}

// The model block contains all probability distributions in the model.
// This of this as specifying the generative model for the scenario.
model {
    real local_t; // declaring extra variables to store intermidiate value, used in computing the seasonal component
    real season; // declaring extra variables to store intermidiate value, used in the mean of the Normal 
    
    // params for trend 
    c1 ~ normal(0,10); //
    c2 ~ normal(0,10);
    //c2 ~ exponential(0.000001);//
    c3 ~ exponential(0.000001);//
    c4 ~ normal(0,10); //
    c5 ~ normal(0,10);
    //c5 ~ exponential(0.000001);//
    c6 ~ exponential(0.000001);//
    
    // params for seasonality
    //A ~ gamma(8.55712644,7.55717524);
    A ~ gamma(11.38643189,37.76884397);
    phi ~ exponential(2/T);
    r ~ exponential(2/T);
    
    // sigma for noise 
    sig ~ exponential(.000001);
    
    // loop through the data points
    for (i in 1:num) {
        // localT in the function for the asymmetric function above
        local_t = fmod(t[i] - phi, T);

        season = (local_t<=r ? -A : A) + (local_t<=r ? 2 : -2)*A*(fmod(t[i]-phi,T)-(local_t<=r ? 0 : r))/(local_t<=r ? (r) : (T-r));
        if (t[i] < 12150)                         
            co2[i] ~ normal(c1 + c2*t[i] + c3*square(t[i]) + season, sig);
        else
            co2[i] ~ normal(c4 + c5*t[i] + c6*square(t[i]) + season, sig);

    }
}

"""
stan_model2 = pystan.StanModel(model_code=stan_code)
stan_results2 = stan_model2.sampling(data=data)
print(stan_results2)
```

```python
stan_code_vani = """

// The data block contains all known quantities - typically the observed
// data and any constant hyperparameters.
data {
    int<lower=1> num;  // number of data in the data set
    real<lower=0> T; // period of seasonality, equals 365.25
    real<lower=0> t[num]; // days since 1958-03-29
    real co2[num]; // co2 level

}

// The parameters block contains all unknown quantities - typically the
// parameters of the model. Stan will generate samples from the posterior
// distributions over all parameters.
parameters {
    real c1; //coefficient in quadratic function 
    real<lower=0> c2; //coefficient in quadratic function 
    real<lower=0> c3; //coefficient in quadratic function 
    real<lower=0> A; // amplitude of the periodic function
    real<lower=0, upper=pi()> phi; // phase
    real<lower=0> sig; // std of the normal noise
    
    
}

// The model block contains all probability distributions in the model.
// This of this as specifying the generative model for the scenario.
model {
    real season; // declaring extra variables to store intermidiate value, used in the mean of the Normal 

    c1 ~ normal(0,10); //
    c2 ~ normal(0,10);
    c3 ~ exponential(0.000001);//

    // params for seasonality
    A ~ gamma(11.38643189,37.76884397);
    phi ~ exponential(2/pi());
    
    // sigma for noise 
    sig ~ exponential(.000001);
    
    // loop through the data points
    for (i in 1:num) {
        season = A*cos((2*pi()*t[i])/365.25 + phi);                      
        co2[i] ~ normal(c1 + c2*t[i] + c3*square(t[i]) + season, sig);

    }
}

"""
stan_model_vani = pystan.StanModel(model_code=stan_code_vani)
stan_results_vani = stan_model_vani.sampling(data=data)
print(stan_results_vani)

```

## Model selection - Evaluate on Test set


```python
import math
parameters_vani = ['c1', 'c2', 'c3', 'A', 'phi', 'sig'] # params of vanilla model
parameters = ['c1', 'c2', 'c3','c4','c5','c6','A', 'phi', 'r','sig']# params of the proposed model
def predict_vani(samples, x_pred):
    """
    Predictions for vanilla model. Assume samples have 4000 samples of the parameters
    
    """
    # x values for 40 years from now 
#     x_end = test_dates[-1] + 365.25*40
#     x_pred = list(range((test_dates[-1], x_end, 7)))

    y_pred = np.zeros((4000, len(x_pred)))
    for i in range(4000):
        c1, c2, c3,A, phi, sig = [samples[key][i] for key in parameters_vani]
        for j in range(len(x_pred)):
            t = x_pred[j]
            y_pred[i,j] = np.random.normal(c1 + c2*t + c3*t**2 + A*np.cos(2*math.pi*t/365.25+phi), sig)


    return y_pred

def predict(samples, x_pred):
    """
    Predictions for the proposed model. Assume samples have 4000 samples of the parameters
    """
    y_pred = np.zeros((4000, len(x_pred)))
    for i in range(4000):
        c1, c2, c3, c4, c5, c6, A, phi, r, sig = [samples[key][i] for key in parameters]
        for j in range(len(x_pred)):
            t = x_pred[j]
            # t is surely to be larger than 12150, so we use the second branch of the piecewise trend
            y_pred[i,j] = np.random.normal(c4 + c5*t + c6*t**2 + periodic(t, phase=phi, T=365.25, amin=-A, amax=A, rising_t=r), sig)
            
    return y_pred



```
